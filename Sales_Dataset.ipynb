```python
# Fill the missing data
df["brand"]= df["brand"].fillna("Unknown")
df["rating"]= df["rating"].fillna(df["rating"].median())
df["number_of_reviews"] = df["number_of_reviews"].fillna(0)
df["location"] = df["location"].fillna("Unknown")
df["stock"]= df["stock"].fillna(0)
df["discount"]= df["discount"].fillna(0)

```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Fix price columns (remove "Rs." and commas convert to number)
df["price"]= df["price"].str.replace("Rs. ", "").str.replace(",", "").astype(float)
df["original_price"] = df["original_price"].str.replace("Rs. ", "").str.replace(",", "").astype(float)
```

```python
df.head()
```

```python
df["shipping_fee"] = df["shipping_fee"].str.replace("Free", "0").str.replace("Rs. ", "").astype(int)
```

Category and Brand analysis

```python
# How many products are in each category?
df['category'].value_counts()
# Which category has the highest average price?
df.groupby("category")["price"].mean().sort_values(ascending=False)

# What is the most expensive and cheapest product?
df.loc[df['price'].idxmax()]
df.loc[df['price'].idxmin()]

# What is the overall average rating?
df['rating'].mean()
# Which brand has the most products?
df['brand'].value_counts()
# Which payment method is most popular?
df["payment_method"].value_counts()

# How many products offer free shipping?
(df["shipping_fee"]==0).sum()
# What is the average warranty period?
df['warranty_months'].mean()
```

```python
# Chart 1: Products per category (Bar)
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='category', order=df['category'].value_counts().index, hue='category')
plt.title('Products per Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Chart 2: Top brands by count (Bar)
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
top_brands = df['brand'].value_counts().head(10)
sns.barplot(x=top_brands.index, y=top_brands.values, palette='magma')
plt.title('Top 10 Brands by Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Chart 3: Avg price per category (Horizontal bar)
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
avg_price = df.groupby('category')['price'].mean()
avg_price.plot(kind='barh', color='skyblue')
plt.title('Avg Price per Category')
plt.xlabel('Average Price (Rs.)')
plt.show()
```

```python
# 5. Availability split (Donut)
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6, 6))
counts = df['availability'].value_counts()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140, colors = ['#ff9999', '#66b3ff', '#99ff99'])
center_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)
plt.title('Availability Split (In Stock vs Out of Stock?)')
plt.show()
```

```python
# Chart 7: Price distribution (Histogram)
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(df['price'].dropna(), bins=20, kde=True, color='purple')
plt.title('Price Distribution')
plt.xlabel('Price (Rs.)')
plt.show()
```

```python
# Chart 8: Payment method usage (Bar)
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='payment_method', order=df['payment_method'].value_counts().index, hue='payment_method')
plt.title('Payment Method Usage')
plt.xticks(rotation=45)
plt.show()
```

```python
# Chart 9: Stock levels per category (Boxplot)
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='category', y='stock', hue='category')
plt.title('Stock Levels per Category')
plt.xticks(rotation=45)
plt.show()
```

```python
#11
import matplotlib.pyplot as plt
import seaborn as sns

# Top revenue brand in each category
top_brands = (
    df.groupby(['category', 'brand'])['price']
    .sum()
    .reset_index(name='revenue')
    .sort_values('revenue', ascending=False)
    .drop_duplicates('category')
)

# Plot
plt.figure(figsize=(14,7))

ax = sns.barplot(
    data=top_brands,
    x='category',
    y='revenue',
    palette='Blues_r'
)

# Labels
for i, row in enumerate(top_brands.itertuples()):
    ax.text(i, row.revenue, f"Rs.{row.revenue:,.0f}", ha='center')
    ax.text(i, row.revenue/2, row.brand, ha='center', color='white')

plt.title('Top Revenue Brand in Each Category')
plt.xticks(rotation=20)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Least revenue brand in each category
least_brands = (
    df.groupby(['category', 'brand'])['price']
    .sum()
    .reset_index(name='revenue')
    .sort_values('revenue')
    .drop_duplicates('category')
)

# Plot
plt.figure(figsize=(14,7))

ax = sns.barplot(
    data=least_brands,
    x='category',
    y='revenue',
    palette='Reds_r'
)

# Labels
for i, row in enumerate(least_brands.itertuples()):
    ax.text(i, row.revenue, f"Rs.{row.revenue:,.0f}", ha='center')
    ax.text(i, row.revenue/2, row.brand, ha='center', color='white')

plt.title('Least Revenue Brand in Each Category')
plt.xticks(rotation=20)

plt.tight_layout()
plt.show()
```

```python
#13
import matplotlib.pyplot as plt
import seaborn as sns

# Revenue by category
category_rev = (
    df.groupby('category')['revenue']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
# Plot
plt.figure(figsize=(14,7))
ax = sns.barplot(
    data=category_rev,
    x='revenue',
    y='category',
    palette='Blues_r'
)
# Revenue labels
for i, row in enumerate(category_rev.itertuples()):
    ax.text(row.revenue, i, f" Rs.{row.revenue:,.0f}", va='center')

plt.title('Highest Revenue Category')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Lowest revenue categories
category_rev = (
    df.groupby('category')['revenue']
    .sum()
    .sort_values()   # Lowest to highest
    .reset_index()
)
# Plot
plt.figure(figsize=(14,7))

ax = sns.barplot(
    data=category_rev,
    x='revenue',
    y='category',
    palette='flare'
)
# Labels
for i, row in enumerate(category_rev.itertuples()):
    ax.text(row.revenue, i, f" Rs.{row.revenue:,.0f}", va='center')

plt.title('Lowest Revenue Categories')
plt.xticks([])

sns.despine(bottom=True)
plt.tight_layout()
plt.show()
```

```python
#14 Highest vs lowest revenue category
import matplotlib.pyplot as plt
import seaborn as sns

# Highest & lowest revenue categories
category_rev = df.groupby('category')['revenue'].sum()

highest_cat = category_rev.idxmax()
lowest_cat = category_rev.idxmin()

# Brand revenue comparison
brand_rev = (
    df[df['category'].isin([highest_cat, lowest_cat])]
    .groupby(['category', 'brand'])['revenue']
    .sum()
    .reset_index()
)

# Plot
plt.figure(figsize=(14,7))

ax = sns.barplot(
    data=brand_rev,
    x='revenue',
    y='brand',
    hue='category',
    palette=['#1f77b4', '#ff7f0e']
)

# Labels
for c in ax.containers:
    ax.bar_label(c, fmt='Rs.%.0f')

plt.title(f'{highest_cat} vs {lowest_cat} Brand Revenue')
plt.tight_layout()
plt.show()
```

```python
#15 top 10 products by revenue
import matplotlib.pyplot as plt
import seaborn as sns
top_products = (
    df.groupby('product_name')['revenue']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)
# Plot
plt.figure(figsize=(14,7))

ax = sns.barplot(
    data=top_products,
    x='revenue',
    y='product_name',
    palette='Blues_r'
)

# Labels
for i, row in enumerate(top_products.itertuples()):
    ax.text(row.revenue, i, f" Rs.{row.revenue:,.0f}", va='center')

plt.title('Top 10 Revenue Products')
plt.tight_layout()
plt.show()
```

```python
#Bottom 10 least revenue generating products
import matplotlib.pyplot as plt
import seaborn as sns

# Bottom 10 products by revenue
bottom_10 = (
    df.groupby('product_name')['revenue']
    .sum()
    .sort_values()
    .head(10)
    .reset_index()
)

# Plot
plt.figure(figsize=(12,7))

ax = sns.barplot(
    data=bottom_10,
    x='revenue',
    y='product_name',
    palette='Reds_r'
)

# Labels
for i, row in enumerate(bottom_10.itertuples()):
    ax.text(row.revenue, i, f" Rs.{row.revenue:,.0f}", va='center')

plt.title('Top 10 Least Revenue Products')
plt.tight_layout()
plt.show()
```

```python
#18.Inventory availaible by category

import matplotlib.pyplot as plt
import seaborn as sns

# Count availability by category
availability = (
    df.groupby(['category', 'availability'])
    .size()
    .reset_index(name='count')
)

# Plot
plt.figure(figsize=(14,7))

ax = sns.barplot(
    data=availability,
    x='category',
    y='count',
    hue='availability',
    palette=['#2ecc71', '#e74c3c', '#f1c40f']
)

# Labels
for p in ax.patches:
    ax.annotate(
        int(p.get_height()),
        (p.get_x() + p.get_width()/2, p.get_height()),
        ha='center',
        xytext=(0,5),
        textcoords='offset points'
    )

plt.title('Inventory Availability by Category')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
