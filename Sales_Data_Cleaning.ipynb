```python
# Fill the missing data
df["brand"]= df["brand"].fillna("Unknown")
df["rating"]= df["rating"].fillna(df["rating"].median())
df["number_of_reviews"] = df["number_of_reviews"].fillna(0)
df["location"] = df["location"].fillna("Unknown")
df["stock"]= df["stock"].fillna(0)
df["discount"]= df["discount"].fillna(0)

```

```python
# Fix price columns (remove "Rs." and commas, convert to number)
df["price"]= df["price"].str.replace("Rs. ", "").str.replace(",", "").astype(float)
df["original_price"] = df["original_price"].str.replace("Rs. ", "").str.replace(",", "").astype(float)
```

```python
df["shipping_fee"] = df["shipping_fee"].str.replace("Free", "0").str.replace("Rs. ", "").astype(int)
```

```python
# How many products are in each category?
df['category'].value_counts()
# Which category has the highest average price?
df.groupby("category")["price"].mean().sort_values(ascending=False)

# What is the most expensive and cheapest product?
df.loc[df['price'].idxmax()]
df.loc[df['price'].idxmin()]

# What is the overall average rating?
df['rating'].mean()
# Which brand has the most products?
df['brand'].value_counts()
# Which payment method is most popular?
df["payment_method"].value_counts()

# How many products offer free shipping?
(df["shipping_fee"]==0).sum()
# What is the average warranty period?
df['warranty_months'].mean()
```

```python
# 11. Revenue Analysis by Brand & Category
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Revenue Analysis by Brand & Category

# Create revenue column
df['revenue'] = df['price']

# Group revenue by category and brand
cat_brand_revenue = (
    df.groupby(['category', 'brand'], as_index=False)['revenue']
    .sum()
)

# Select top brand from each category
top_brand_each_category = (
    cat_brand_revenue
    .sort_values('revenue', ascending=False)
    .groupby('category')
    .head(1)
    .sort_values('revenue', ascending=False)
)


# Visualization
plt.figure(figsize=(14,7))

ax = sns.barplot(
    data=top_brand_each_category,
    x='category',
    y='revenue'
)

# Add revenue labels
for i, row in enumerate(top_brand_each_category.itertuples()):
    ax.text(
        i,row.revenue,f"Rs. {row.revenue:,.0f}",ha='center',va='bottom',fontsize=10
    )

# Brand labels inside bars
for i, row in enumerate(top_brand_each_category.itertuples()):
    ax.text(i,row.revenue * 0.5,row.brand,ha='center',fontsize=1)

# Styling
plt.title(
    'Top Revenue-Generating Brand in Each Category',
    fontsize=16,
    weight='bold'
)

plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Revenue', fontsize=12)

plt.xticks(rotation=20, fontsize=10)

sns.despine()

plt.tight_layout()
plt.show()
```

```python
#12. Top 10 lowest generated revenue
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate total revenue per brand
brand_revenue = df.groupby('brand')['price_numeric'].sum().reset_index()
brand_revenue.columns = ['Brand', 'Total Revenue']

# Get 10 least revenue generating brands
# Filtering out brands with 0 revenue if they exist
bottom_10_brands = brand_revenue[brand_revenue['Total Revenue'] > 0].sort_values(by='Total Revenue').head(10)

# Step 5: Plotting
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# Create the bar plot
ax = sns.barplot(
    data=bottom_10_brands,
    x='Total Revenue',
    y='Brand',
    palette='Reds_r' # Red palette for "bottom" or "low" metrics
)

# Add data labels
for i in ax.containers:
    ax.bar_label(i, padding=3, fmt='Rs. {:,.0f}', fontsize=10, weight='bold')

# Formatting labels and title
plt.title('Top 10 Brands with Lowest Generated Revenue', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Total Revenue (in Rs.)', fontsize=12, fontweight='bold')
plt.ylabel('Brand Name', fontsize=12, fontweight='bold')

# Refine layout
plt.tight_layout()
plt.savefig('least_revenue_brands.png')

# Display the data used for the graph in the output
print(bottom_10_brands)
```

```python
#14. Highest vs Lowest Revenue Category
# Brand Performance Comparison

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Revenue by category
category_rev = (
    df.groupby('category', as_index=False)['revenue']
    .sum()
)

# Identify highest and lowest categories
highest_cat = category_rev.loc[
    category_rev['revenue'].idxmax(),
    'category'
]

lowest_cat = category_rev.loc[
    category_rev['revenue'].idxmin(),
    'category'
]

# Filter only those categories
filtered = df[df['category'].isin([highest_cat, lowest_cat])]

# Brand revenue comparison
brand_rev = (
    filtered.groupby(['category', 'brand'], as_index=False)['revenue']
    .sum()
    .sort_values('revenue', ascending=True)
)

# -----------------------------------
# Visualization
# -----------------------------------

sns.set_style("whitegrid")

plt.figure(figsize=(14,7))

# Professional color palette
palette = {
    highest_cat: '#1f77b4',   # blue
    lowest_cat: '#ff7f0e'     # orange
}

# Horizontal grouped barplot
ax = sns.barplot(
    data=brand_rev,
    x='revenue',
    y='brand',
    hue='category',
    palette=palette
)

# Add revenue labels
for container in ax.containers:
    ax.bar_label(
        container,
        fmt=' Rs. %.0f',
        padding=3,
        fontsize=9
    )

# Titles and labels
plt.title(
    f'Brand Revenue Comparison: {highest_cat} vs {lowest_cat}',
    fontsize=17,
    weight='bold',
    pad=20
)

plt.xlabel(
    'Revenue',
    fontsize=13
)

plt.ylabel(
    'Brand',
    fontsize=13
)

# Legend styling
plt.legend(
    title='Category',
    fontsize=10,
    title_fontsize=11
)

# Improve readability
plt.xticks(fontsize=11)
plt.yticks(fontsize=10)

# Remove extra borders
sns.despine(left=False, bottom=False)

# Improve spacing
plt.subplots_adjust(left=0.25, right=0.95)

plt.show()
```

```python
#18.Inventory availaible by category

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# 3. Aggregate data by Category and Availability
availability_counts = df.groupby(['category', 'availability']).size().reset_index(name='count')

# 4. Professional Visualization Setup
sns.set_theme(style="whitegrid")
plt.figure(figsize=(14, 8))

# Define a professional color palette
# Green for In Stock, Red for Out of Stock, Yellow/Orange for Limited
status_colors = {"In Stock": "#2ecc71", "Out of Stock": "#e74c3c", "Limited": "#f1c40f"}

# Create grouped bar chart
barplot = sns.barplot(
    data=availability_counts,
    x='category',
    y='count',
    hue='availability',
    palette=status_colors,
    edgecolor='black',
    linewidth=0.5
)

# 5. Styling and Labels
plt.title('Inventory Availability Status by Category', fontsize=18, fontweight='bold', pad=25)
plt.xlabel('Product Category', fontsize=13, fontweight='semibold')
plt.ylabel('Total Product Count', fontsize=13, fontweight='semibold')
plt.xticks(rotation=45)
plt.legend(title='Stock Status', title_fontsize='12', loc='upper right', frameon=True)

# 6. Add Data Labels (Values on top of bars)
for p in barplot.patches:
    height = p.get_height()
    if height > 0:
        barplot.annotate(f'{int(height)}',
                         (p.get_x() + p.get_width() / 2., height),
                         ha='center', va='center',
                         xytext=(0, 10),
                         textcoords='offset points',
                         fontsize=10, fontweight='bold')

# 7. Final Adjustments and Export
plt.tight_layout()
plt.savefig('availability_by_category.png', dpi=300)

# 8. Create a summary table for reference
availability_pivot = availability_counts.pivot(index='category', columns='availability', values='count').fillna(0)
availability_pivot.to_csv('availability_summary_by_category.csv')
```
