Step 4: Checking for any missing data

```python
df.isnull().sum()
```

Step 5: Filling the missing data

```python
x=df["torque_nm"].median()
df.fillna({"torque_nm":x},inplace=True)

```

```python
x=df["towing_capacity_kg"].mean()
df.fillna({"towing_capacity_kg":x},inplace=True)
```

```python
x=df["fast_charging_power_kw_dc"].mode()
df.fillna({"fast_charging_power_kw_dc":x},inplace=True)
```

```python
df.columns = df.columns.str.strip()
x=df["fast_charge_port"].mode()
df.fillna({"fast_charge_port":x},inplace=True)
```

```python
df.info()
```

Step 6: Remove Duplicate Data

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

Step-9: Removing Irrelevant Columns

```python
df.drop(columns=['number_of_cells'],inplace=True)
df.drop(columns=['source_url'],inplace=True)
```

```python
df["brand"].nunique()
# Brand with highest number of model
df["brand"].value_counts().idxmax()    #value_counts: counts how many models each brand has   idxmax:returns the brand with the highest count
# What is the distribution of car segment
df['segment'].value_counts()
# Which drivetrain is most common
df['drivetrain'].value_counts().idxmax()
# Average range of all electric cars
df['range_km'].mean()
#Median battery capacity
df['battery_capacity_kWh'].median()
# Which car body type appears the most
df['car_body_type'].mode()
# Average sitting capacity
df['seats'].mean()
```
